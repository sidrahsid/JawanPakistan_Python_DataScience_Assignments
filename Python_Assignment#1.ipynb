**P#3 Program to print Current date and Time**

```python
from datetime import datetime

current = datetime.now()
 
print("Current date time =", current)

# formatting date and time as:
cdt = current.strftime("%d/%m/%Y %I:%M:%S")
print("Current date and time =", cdt)	
```

**P#4 Program to COMPUTE AREA OF CIRCLE**

```python
import math
#take input as radius'r' from user
r=float(input("Enter Radius of Circle"))
area=(math.pi)*(r*r)
print("Area of Circle= ",area)
```

**P#5 Program to Print First and Last name of user in revrse order**

```python
fname = input("Enter your First Name : ")
lname = input("Enter your Last Name : ")
print ("Hi !  " + lname + " " + fname)

# OR REVERSING STRINGS WE GET
fname = fname[::-1]
lname = lname[::-1]
print ("Hi !  " + lname + " " + fname)
```
